# t-SNE

```python
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data
df = pd.read_excel('data_10k.xlsx')

# Drop the target column
X = df.drop(columns=['P_NED_ADR'])

# Apply t-SNE for 2D reduction
tsne_2d = TSNE(n_components=2, random_state=42)
X_2d = tsne_2d.fit_transform(X)

# Plot 2D heatmap
plt.figure(figsize=(10, 8))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c='blue', cmap='viridis', s=5, alpha=0.7)
plt.colorbar()
plt.title('2D t-SNE Heatmap')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.show()


```

```python
# Apply t-SNE for 3D reduction
tsne_3d = TSNE(n_components=3, random_state=42)
X_3d = tsne_3d.fit_transform(X)

# Plot 3D heatmap
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_3d[:, 0], X_3d[:, 1], X_3d[:, 2], c='blue', cmap='viridis', s=5, alpha=0.7)
fig.colorbar(scatter, ax=ax)
ax.set_title('3D t-SNE Heatmap')
ax.set_xlabel('t-SNE 1')
ax.set_ylabel('t-SNE 2')
ax.set_zlabel('t-SNE 3')
plt.show()
```

```python
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data
df = pd.read_excel('data_10k.xlsx')

# Separate features and target
X = df.drop(columns=['P_NED_ADR'])
y = df['P_NED_ADR']

# Apply t-SNE for 2D reduction
tsne_2d = TSNE(n_components=2, random_state=42)
X_2d = tsne_2d.fit_transform(X)

# Plot 2D heatmap with labels
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y, cmap='viridis', s=5, alpha=0.7)
plt.colorbar(scatter, label='P_NED_ADR')
plt.title('2D t-SNE Heatmap')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.savefig('tsne_2d.pdf')
plt.show()


```

```python
# Plot 2D heatmap with labels
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y, s=5, alpha=0.7)
plt.colorbar(scatter, label='P_NED_ADR')
plt.title('2D t-SNE Heatmap')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.savefig('tsne_2d.pdf')
plt.show()
```

```python
# Define colors for the labels
colors = {0: 'blue', 1: 'red'}
label_names = {0: 'Class 0', 1: 'Class 1'}

# Plot 2D heatmap with labels
plt.figure(figsize=(8, 8))
for label in colors:
    indices = y == label
    plt.scatter(X_2d[indices, 0], X_2d[indices, 1], c=colors[label], label=label_names[label], s=5, alpha=0.7)
plt.legend()
plt.title('2D t-SNE Heatmap')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.tight_layout()
plt.savefig('tsne_2d.pdf')
plt.show()
```

```python
# Plot 3D heatmap with labels
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
for label in colors:
    indices = y == label
    ax.scatter(X_3d[indices, 0], X_3d[indices, 1], X_3d[indices, 2], c=colors[label], label=label_names[label], s=5, alpha=0.7)
ax.legend()
ax.set_title('3D t-SNE Heatmap')
ax.set_xlabel('t-SNE 1')
ax.set_ylabel('t-SNE 2')
ax.set_zlabel('t-SNE 3')
plt.tight_layout()
plt.savefig('tsne_3d.pdf')
plt.show()
```

# PCA

```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the data
df = pd.read_excel('data_10k.xlsx')

# Separate features and target
X = df.drop(columns=['P_NED_ADR'])
y = df['P_NED_ADR']

# Define colors for the labels
colors = {0: 'blue', 1: 'red'}
label_names = {0: 'Class 0', 1: 'Class 1'}

# Apply PCA for 2D reduction
pca_2d = PCA(n_components=2)
X_2d = pca_2d.fit_transform(X)

# Plot 2D heatmap with labels
plt.figure(figsize=(8, 8))
for label in colors:
    indices = y == label
    plt.scatter(X_2d[indices, 0], X_2d[indices, 1], c=colors[label], label=label_names[label], s=5, alpha=0.7)
plt.legend()
plt.xlim(-10,400)
plt.ylim(-10,300)
plt.title('2D PCA Heatmap')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.tight_layout()
plt.savefig('PCA_2d.pdf')
plt.show()
```

```python
# Apply PCA for 3D reduction
pca_3d = PCA(n_components=3)
X_3d = pca_3d.fit_transform(X)

# Plot 3D heatmap with labels
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
for label in colors:
    indices = y == label
    ax.scatter(X_3d[indices, 0], X_3d[indices, 1], X_3d[indices, 2], c=colors[label], label=label_names[label], s=5, alpha=0.7)
ax.legend()
ax.set_title('3D PCA Heatmap')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_xlim(-10, 500)
ax.set_ylim(-10, 450)
ax.set_zlim(-10, 450)
plt.tight_layout()
plt.savefig('PCA_3d.pdf')
plt.show()
```

# LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt
import numpy as np

# Load the data
df = pd.read_excel('data_10k.xlsx')

# Separate features and target
X = df.drop(columns=['P_NED_ADR'])
y = df['P_NED_ADR']

# Define colors for the labels
colors = {0: 'blue', 1: 'red'}
label_names = {0: 'Class 0', 1: 'Class 1'}

# Apply LDA for 1D reduction (since we have 2 classes)
lda = LDA(n_components=1)
X_lda = lda.fit_transform(X, y)

# Since we have only 1 component, we need to manually create 2D data for plotting
X_lda_2d = X_lda if X_lda.shape[1] == 2 else np.hstack((X_lda, np.zeros((X_lda.shape[0], 1))))

# Plot 2D LDA heatmap with labels
plt.figure(figsize=(10, 8))
for label in colors:
    indices = y == label
    plt.scatter(X_lda_2d[indices, 0], X_lda_2d[indices, 1], c=colors[label], label=label_names[label], s=5, alpha=0.7)
plt.legend()
plt.title('2D LDA Heatmap')
plt.xlabel('LDA Component 1')
plt.ylabel('Dummy Component (Zero)')
plt.show()
```

```python
from sklearn.manifold import MDS

# Apply MDS for 2D reduction
mds = MDS(n_components=2, random_state=42)
X_mds = mds.fit_transform(X)

# Plot 2D MDS heatmap with labels
plt.figure(figsize=(10, 8))
for label in colors:
    indices = y == label
    plt.scatter(X_mds[indices, 0], X_mds[indices, 1], c=colors[label], label=label_names[label], s=5, alpha=0.7)
plt.legend()
plt.title('2D MDS Heatmap')
plt.xlabel('MDS Component 1')
plt.ylabel('MDS Component 2')
plt.show()
```
