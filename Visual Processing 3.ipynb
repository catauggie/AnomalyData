```python
import pandas as pd
df = pd.read_excel('resilts_MLP_recall.xlsx')
df = df.sort_values(by='val_precision')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Строим график
plt.figure(figsize=(15, 8))

epochs = np.arange(df.shape[0])

# Кривые для validation
plt.plot(epochs, df['val_accuracy'], marker='o', linestyle='-', color='blue', label='val_accuracy')
plt.plot(epochs, df['val_precision'], marker='o', linestyle='-', color='orange', label='val_precision')
plt.plot(epochs, df['val_recall'], marker='o', linestyle='-', color='green', label='val_recall')

# Кривые для test
plt.plot(epochs, df['test_accuracy'], marker='o', linestyle='-', color='red', label='test_accuracy')
plt.plot(epochs, df['test_precision'], marker='o', linestyle='-', color='purple', label='test_precision')
plt.plot(epochs, df['test_recall'], marker='o', linestyle='-', color='brown', label='test_recall')

# Настройки графика
plt.title('Metrics Evolution')
plt.xlabel('Individual')
plt.ylabel('Score')
plt.grid(True)
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Пример данных (замените на свой DataFrame)
# Пусть df уже содержит данные по layers

# Строим график
plt.figure(figsize=(15, 8))

epochs = np.arange(df.shape[0])

# Кривая для layers
plt.scatter(epochs, df['layers'], marker='<', color='blue', label='layers')

# Добавление фона с номерами
plt.axvspan(0, 19, color='gray', alpha=0.1, label='Generation 0')

for i in range(20, df.shape[0], 10):
    plt.axvspan(i, i + 9, color=plt.cm.tab20(i // 10), alpha=0.1, label=f'Generation {i // 10}')

# Настройки графика
plt.title('Evolution of Neural Network Layers')
plt.xlabel('Individuals')
plt.ylabel('Layers Configuration')
plt.grid(True)
plt.legend()

# Отображение графика
plt.tight_layout()
plt.savefig('layers_plot.png')
plt.show()
```

```python
# Строим график
plt.figure(figsize=(15, 8))

epochs = np.arange(df.shape[0])

# Создаем словарь для соответствия маркеров функциям активации
activation_markers = {
    'identity': 'o',
    'logistic': 's',
    'tanh': '^',
    'relu': 'D'
}

# Кривая для layers с разными маркерами по функции активации
for activation, marker in activation_markers.items():
    mask = df['activation'] == activation
    plt.scatter(epochs[mask], df[mask]['layers'], marker=marker, label=f'Activation: {activation}')

# Добавление фона с номерами
plt.axvspan(0, 19, color='gray', alpha=0.1, label='Generation 0')

for i in range(20, df.shape[0], 10):
    plt.axvspan(i, i + 9, color=plt.cm.tab20(i // 10), alpha=0.1, label=f'Generation {i // 10}')
    
# Добавление фона с номерами
#plt.axvspan(0, df.shape[0], color='black', alpha=0.1)

# Добавление фона с номерами
for i in range(0, df.shape[0], 2):
    plt.axvspan(i, i + 1, color='gray', alpha=0.1)


        
# Настройки графика
plt.title('Evolution of Neural Network Layers')
plt.xlabel('Individuals')
plt.ylabel('Layers Configuration')
plt.grid(True)

# Добавляем легенду
plt.legend()

# Отображение графика
plt.tight_layout()
plt.savefig('layers_plot_with_activation.pdf')
plt.show()
```
